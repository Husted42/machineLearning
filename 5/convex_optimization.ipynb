# Exercse ii

### Functions

```python
# Define the function f(x_1, x_2)
def f(x):
    x1, x2 = x
    return np.exp(x1 + 4*x2 - 0.3) + np.exp(x1 - 0.3) + np.exp(-x1 - 0.3)

# I have manually comoputed gradient of f(x_1, x_2), this is defined as grad_f(x)
def grad_f(x):
    x1, x2 = x
    df_dx1 = np.exp(x1 + 4*x2 - 0.3) + np.exp(x1 - 0.3) - np.exp(-x1 - 0.3)
    df_dx2 = 4 * np.exp(x1 + 4*x2 - 0.3)
    return np.array([df_dx1, df_dx2])

# Follow the gradient, and hopefully we will reach near the optimal point
# x_0 will be the starting point
# grad is the gradient of f(x_1, x_2)
# aplha is the learning rate
# tol is the tolerance, if the difference between two consecutive points is less than tol, we will stop
# max_iter is the maximum number of iterations
def gradient_descent(x0, grad, alpha, tol=1e-6, max_iter=1000):
    x = x0
    convergence_curve = []
    error_curve = []

    for t in range(max_iter):
        x_prev = x #Store x as the previous point
        x = x - alpha(t) * grad(x) # Compute the next point
        error = np.linalg.norm(x - optimal_point ) 
        f_error = f(x) - f(optimal_point )

        convergence_curve.append(f_error)
        error_curve.append(error)

        if np.linalg.norm(x - x_prev) < tol:
            break

    return x, convergence_curve, error_curve, t
```

### Variables

```python
alpha = lambda t: 1 / (t + 1)  # Step size
optimal_point  = np.array([-0.5*np.log(2), 0])  # Optimal point
x0 = np.array([0, 0])  # Starting point
```

### Compute and evaluate

```python
# Perform gradient descent
x_opt, convergence_curve, error_curve, number_of_iteration = gradient_descent(x0, grad_f, alpha)

print("Optimal point: ", x_opt)
print("Function value: ", f(x_opt))
print("Number of iterations: ", number_of_iteration)
print("Error: ", error_curve[-1])
print("Convergence: ", convergence_curve[-1]) # How close do we get to the optimal point
```

```python
# Plotting the convergence curves
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(convergence_curve, label='f(x_t) - f(x^*)')
plt.xlabel('Iteration')
plt.xlim(0, 20)
plt.title('Convergence of Objective Function')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(error_curve, label='||x_t - x^*||')
plt.xlabel('Iteration')
plt.title('Convergence of x_t to x^*')
plt.xlim(0, 20)
plt.legend()

plt.tight_layout()
plt.show()
```
