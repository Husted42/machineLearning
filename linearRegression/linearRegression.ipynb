# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load data
data_file_path = "PCB.dt"
data_matrix = np.loadtxt(data_file_path)
```

# Data

```python
X_, Y_ = [], []
for elm in data_matrix:
    X_.append(elm[0])
    Y_.append(elm[1])
X, Y = np.array(X_), np.array(Y_)
Y = Y.reshape(28,1)
foo = np.ones_like(X).reshape(28, 1)
X = X.reshape(28, 1)
X = np.hstack((X, foo))
X = X.reshape(28, 2)
print(X.T)
print(Y.T)
```

```python
plt.scatter(Y_, X_)
```

# Model 1

```python
def linreg(X, y):
    w = np.dot(X.T, X)
    w = np.linalg.inv(w)
    w = np.dot(w, X.T)
    w = np.dot(w, y)
    return w

linreg(X, Y)
```

```python
#PLot the funtion
def plot1(a, b, y_val, x_val):
    x = np.linspace(0, 30, 50)
    y = (a) * x + (b)
    plt.plot(x, y, color = 'red')
    plt.scatter(x_val, y_val)
    plt.xlim(0, 15)
    plt.show()
plot1(1.55777052, -1.45194395, Y_, X_)
```

# Model 2

```python
def affLin(X, y):
    y = np.log(y)
    return linreg(X, y)
parameters = affLin(X, Y)
print(parameters)

def plot2(a, b, y_val, x_val):
    x = np.linspace(0, 15, 100)
    y = (a) * x + (b)
    plt.plot(x, y, color = 'red')
    plt.scatter(x_val, y_val)
    plt.xlabel('Age')
    plt.ylabel('log(PCB)')
    plt.xlim(0, 15)
    plt.ylim(-2, 5)
    plt.show()
plot2(0.25912824, 0.03147247, np.log(Y), X_)
```

```python
def h(x):
    return np.exp(parameters[0] * x + parameters[1])

def MSE(x, y):
    y = y.reshape(28,)

    vfunc = np.vectorize(h)
    pred = vfunc(x)
    pred = pred.reshape(28,)
    diff = y - pred
    mse = (diff.dot(diff)) / len(y)
    return mse

MSE(X_, Y)

```

```python
# r^2 = 1-MSE/var(y)
r_sqr = 1-(MSE(X_, Y))/(np.var(Y))
print(r_sqr)
```

# Model 3

```python
def nonLin(X, y):
    X.T[0] = np.sqrt(X.T[0])
    y = np.log(y)
    print(X.shape)
    return linreg(X, y)

parameters = nonLin(X, Y)

def plot3(a, b, x_val, y_val):
    x = np.linspace(0, 15, 100)
    y = (a) * x + (b)
    plt.plot(x, y, color = 'red')
    plt.scatter(x_val, y_val)
    plt.xlim(0, 4)
    plt.ylim(-2, 5)
    plt.xlabel('sqrt(Age)')
    plt.ylabel('log(PCB)')
    plt.show()
plot3(parameters[0], parameters[1], np.sqrt(X_), np.log(Y))
print(parameters)

def plot4(a, b, x_val, y_val):
    x = np.linspace(0, 15, 100)
    y = (a) * np.sqrt(x) + (b)
    plt.plot(x, y, color = 'red')
    plt.scatter(x_val, y_val)
    plt.xlim(0, 15)
    plt.ylim(-2, 5)
    plt.xlabel('Age')
    plt.ylabel('log(PCB)')
    plt.show()
plot4(parameters[0], parameters[1], X_, np.log(Y))
```
